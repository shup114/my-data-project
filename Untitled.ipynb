```python
def lvl(r):
    if r >= 250:
        return "特大暴雨"
    elif r >= 100:
        return "大暴雨"
    elif r >= 50:
        return "暴雨"
    elif r >= 25:
        return "大雨"
    elif r >= 10:
        return "中雨"
    elif r >= 0.1:
        return "小雨"
    else:
        return "无降水"

with open("rain.txt", "r", encoding="utf-8") as f, open("out.txt", "w", encoding="utf-8") as o:
    o.write("index\rain24\t降水等级\n")
    next(f)  
    for line in f:
        t = line.strip().split()
        if not t:
            continue
        o.write(f"{t[0]}\t{t[1]}\t{lvl(float(t[1]))}\n")

print("ok")
```

```python
#2026年4月24日作业
def main():
    stations = {}

    with open("station.txt", "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith("#"):
                continue  

            parts = line.split( )
            if len(parts) != 7:
                continue  

            try:
                station_id = parts[0]
                lat = float(parts[1])
                lon = float(parts[2])
                year = int(parts[3])
                month = int(parts[4])
                day = int(parts[5])
                temperature = int(parts[6]) / 10
            except (ValueError, IndexError):
                continue

            if temperature < -100 or temperature > 60:
                continue

            if month not in (12, 1, 2):
                continue

            date_str = f"{year}-{month:02d}-{day:02d}"

            if station_id not in stations:
                stations[station_id] = {
                    "lat": lat,
                    "lon": lon,
                    "temps": [],
                    "dates": []
                }

            stations[station_id]["temps"].append(temperature)
            stations[station_id]["dates"].append(date_str)

    print("stationID,lat,lon,winter_avg_temp,max_temp_date,min_temp_date")
    for sid, info in stations.items():
        temps = info["temps"]
        dates = info["dates"]
        if not temps:
            continue  

        avg_temp = sum(temps) / len(temps)
        max_temp = max(temps)
        min_temp = min(temps)
        max_idx = temps.index(max_temp)
        min_idx = temps.index(min_temp)
        max_date = dates[max_idx]
        min_date = dates[min_idx]

        print(f"{sid},{info['lat']},{info['lon']},{avg_temp:.2f},{max_date},{min_date}")


if __name__ == "__main__":
    main()
```

```python
#2026年4月28日作业
class Station:
    def __init__(self, station_id, lat, lon):
        self.station_id = station_id
        self.lat = lat
        self.lon = lon
        self.temps = []
        self.dates = []

    def add_record(self, temperature, date_str):
        self.temps.append(temperature)
        self.dates.append(date_str)

    def winter_avg_temp(self):
        if not self.temps:
            return None
        return sum(self.temps) / len(self.temps)

    def max_temp_info(self):
        if not self.temps:
            return None, None
        idx = self.temps.index(max(self.temps))
        return self.temps[idx], self.dates[idx]

    def min_temp_info(self):
        if not self.temps:
            return None, None
        idx = self.temps.index(min(self.temps))
        return self.temps[idx], self.dates[idx]


class WeatherData:
    def __init__(self):
        self.stations = {}  # key: station_id, value: Station 对象

    def load_file(self, filename):
        with open(filename, "r", encoding="utf-8") as f:
            for line in f:
                line = line.strip()
                if not line or line.startswith("#"):
                    continue
                parts = line.split()  
                if len(parts) != 7:
                    continue
                try:
                    station_id = parts[0]
                    lat = float(parts[1])
                    lon = float(parts[2])
                    year = int(parts[3])
                    month = int(parts[4])
                    day = int(parts[5])
                    temperature = int(parts[6]) / 10
                except (ValueError, IndexError):
                    continue
                if temperature < -100 or temperature > 60:
                    continue
                if month not in (12, 1, 2):
                    continue
                date_str = f"{year}-{month:02d}-{day:02d}"

                if station_id not in self.stations:
                    self.stations[station_id] = Station(station_id, lat, lon)

                self.stations[station_id].add_record(temperature, date_str)

    def print_summary(self):
        print("stationID,lat,lon,winter_avg_temp,max_temp_date,min_temp_date")
        for sid, station in self.stations.items():
            avg_temp = station.winter_avg_temp()
            max_temp, max_date = station.max_temp_info()
            min_temp, min_date = station.min_temp_info()
            if avg_temp is not None:
                print(f"{sid},{station.lat},{station.lon},{avg_temp:.2f},{max_date},{min_date}")


if __name__ == "__main__":
    weather = WeatherData()
    weather.load_file("station.txt")
    weather.print_summary()
```
